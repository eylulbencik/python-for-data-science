```python
# multi index
```

```python
first_index = ["bencik","bencik","bencik","atici","atici","atici"]
```

```python
second_index = ["ela","eylul","umit","yagiz","yusuf","nilufer"]
```

```python
zipped_index = list(zip(first_index,second_index))
zipped_index
```

```python
zipped_index = pd.MultiIndex.from_tuples(zipped_index)
```

```python
zipped_index
```

```python
sample_values = np.ones((6,3))
```

```python
big_df = pd.DataFrame(sample_values, index = zipped_index, columns = ["age","gender","birth year"])
```

```python
big_df
```

```python
big_df.loc["bencik"].loc["ela"]
```

```python
# for changing values

# Just updates "bencik" -> "ela" 
big_df.loc[("bencik", "ela"), :] = [24, "female", "2002"]
```

```python
# Z-Score
```

```python
data = np.array([10,12,13,15,21,100,104])
```

```python
# standard deviation is typically consistent with the mean,
# the presence of outliers (100, 104) significantly skews the results.
```

```python
mean = np.mean(data)
print(mean)
```

```python
print(data.sum() / len(data))
```

```python
std = np.std(data)
print(std)
```

```python
# z scores
```

```python
z_score = (data - mean) / std
z_score
```

```python
# outliers

z_score[z_score > 1]
```

```python
data[z_score > 1]
```
